```python
def display_board(board):
    print(board[1]+' | '+board[2]+' | '+board[3])
    print(board[4]+' | '+board[5]+' | '+board[6])
    print(board[7]+' | '+board[8]+' | '+board[9])
```

```python
def player_mark():
    p1=input('choose your mark X/O')
    if p1=="x":
        p2='o'
    else:
        p2='x'
    return [p1,p2]
```

```python
p1,p2=player_mark()
```

```python
def place_mark_in_board(board,mark,position):
    board[position]=mark
```

```python
def check_win(board,mark):
    return ((board[1]==board[2]==board[3]==mark)
    or (board[4]==board[5]==board[6]==mark)
    or (board[7]==board[8]==board[9]==mark)
    or (board[1]==board[4]==board[7]==mark)
    or (board[2]==board[5]==board[8]==mark)
    or (board[3]==board[6]==board[9]==mark)
    or (board[1]==board[5]==board[9]==mark)
    or (board[3]==board[5]==board[1]==mark))
```

```python

board=['']*10
display_board(board)
```

```python

def TicTac():
    board=['']*10
    p1='x'
    while True:
        
        print("p1 makes your choise")
        loc1=choose_your_loc()
        clear_output(wait=True)
        place_mark_in_board(board,p1,loc1)
        
        print("p2 makes your choise")
        loc2=choose_your_loc()
        place_mark_in_board(board,p2,loc2)
        display_board(board)


        if check_win(board,p1):
            print("p1 you win !")
            break
        elif check_win(board,p2):
            print("p2 you win !")
            break
        else:
            continue
```
